```python
import pandas as pd
import numpy as np 
```

```python
col_names = ["PassengerId","Survived","Pclass","Name","Sex","Age","SibSp","Parch","Ticket","Fare","Cabin","Embarked"]

data = pd.read_csv("train.csv",header=None,names=col_names)
data
```

```python
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()
#Encode labels in the columns
data["Sex"] = label_encoder.fit_transform(data["Sex"])
data["Embarked"] = label_encoder.fit_transform(data["Embarked"])
data["Cabin"] = label_encoder.fit_transform(data["Cabin"])

# y= label_encoder.fit_transform(y)
data
```

```python
X = data.values[1:,[2,4,5,6,7,9,11]]
y = data.values[1:,1]
y
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(X[:,:])
X[:,:] = imputer.transform(X[:,:])
```

```python
from sklearn.tree import DecisionTreeClassifier
clf_entropy = DecisionTreeClassifier(criterion='entropy',max_depth=2)
clf_entropy = clf_entropy.fit(X,y)
```

```python
testdata = pd.read_csv("test.csv")
testdata = testdata.iloc[:,[1,3,4,5,6,8,10]]
testdata
```

```python
testdata["Sex"] = label_encoder.fit_transform(testdata["Sex"])
testdata["Embarked"] = label_encoder.fit_transform(testdata["Embarked"])
# testdata["Cabin"] = label_encoder.fit_transform(testdata["Cabin"])
testdata
```

```python
ppf = testdata.values[:,:]
ppf
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
imputer.fit(ppf[:,:])
ppf[:,:] = imputer.transform(ppf[:,:])
```

```python
y_pred = clf_entropy.predict(ppf)
y_pred
```

```python
xyz = pd.read_csv("gender_submission.csv",header=None)
xyz = xyz.iloc[1:,1]
xyz
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(xyz, y_pred)
print(cm)
```
